```
import pandas as pd
from ntscraper import Nitter
```

```
scraper = Nitter()
```

```
tweet = []
user = ['imVkohli']

tmp_user = user
for _ in range(5):
  member = tmp_user.pop()
  tweet_number = scraper.get_profile_info(member)['stats']['tweets']

  tweets = scraper.get_tweets(member , mode='user', number=tweet_number)

  for tweet in tweets['tweets']:
    data = [member, tweet['text'], tweet['date']]
    tweet.append(data)

    if tweet['is_retweet']:
      user.append(tweet['user']['username'][1:])
```

```
def get_tweets(name, modes, no):
  tweets = scraper.get_tweets(name , mode=modes, number=no)

  final_tweet = []

  for tweet in tweets['tweets']:
    data = [tweet['link'], tweet['text'], tweet['date'], tweet['stats']['likes'], tweet['stats']['comments']]
    final_tweet.append(data)

  data = pd.DataFrame(final_tweet, columns=['link', 'text', 'date', 'likes_no', 'tweets_no'])

  return data
```

```
data = get_tweets('imVkohli', 'user', 40)
data
```

```
data = get_tweets('RishiSunak', 'hashtag', 10)
data
```

```
data = get_tweets('World cup 2023', 'term', 20)
data
```
